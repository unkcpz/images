# Spin precession

```julia
using QuantumOptics
using PyPlot
```

```julia
# Parameters
Ω = 2
t = [0:0.01:10;];
```

```julia
# Hamiltonian
b = SpinBasis(1//2)
H = Ω*sigmaz(b);
```

```julia
# ψ₀ = spindown(b) ⊗ spinup(b)
ψ₀ = spindown(b) + spinup(b)
tout, ψₜ = timeevolution.schroedinger(t, ψ₀, H);
exp_val_z = expect(sigmaz(b), ψₜ);
exp_val_x = expect(sigmax(b), ψₜ);
exp_val_y = expect(sigmay(b), ψₜ);
```

```julia
figure(figsize=(6, 3))
plot(tout, exp_val_z, label="expect z")
plot(tout, exp_val_x, label="expect x")
plot(tout, exp_val_y, label="expect y")
legend()
show()
```
